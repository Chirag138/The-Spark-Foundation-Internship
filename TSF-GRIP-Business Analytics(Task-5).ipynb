# Visualizing features of the data against the Profit

```python
fig , ax = plt.subplots(figsize=(20,15))
sns.countplot(data=df,x='City',order=df['City'].value_counts()[:10].index)
```

The above graph shows the number of orders that has been shipped from top 10 cities

```python
sns.pairplot(data=df)
```

```python
df['Category'].value_counts()
```

```python

fig , ax = plt.subplots(figsize=(8,9))
sns.countplot(data=df,x='Category',palette="Greens_d")
```

```python
df1 = df.groupby('Category')['Profit'].sum()
```

```python
ax = df1.plot(kind='bar',alpha=0.75, rot=0,color=['C0', 'C1', 'C9'],figsize=(12,10))
ax.set_ylabel('Profit')
```

The bar plot gives the profit made on different categories that are shipped to the customer.

```python
df['Sub-Category'].value_counts()
```

```python
fig , ax = plt.subplots(figsize=(15,12))
sns.countplot(data=df,x='Sub-Category')
```

The countplot above plots the sub-category features against the number of times they were shipped to the customers.

```python
df2 = df.groupby('Sub-Category')['Profit'].sum()
ax = df2.plot(kind='bar',alpha=0.75, rot=0,figsize=(17,14))
ax.set_ylabel('Profit')
```

The bar graph shows the profit made by each sub-category item bought by the customer.

Using BOXEN plot visualization

```python
fig,ax=plt.subplots(figsize=(15,10))
sns.boxenplot(x='Category',y='Profit',data=df)
```

```python
df3 = df.groupby('City')['Profit'].sum()
ax = df3.sort_values(ascending=False)[:20].plot(kind='bar',alpha=0.75, rot=0,figsize=(19,14))
ax.set_ylabel('Profit')
```

In the above plot, we visualize the profit made by the store in various cities.

We are done with cities now we are visualizing the states against orders and profit.

```python

fig , ax = plt.subplots(figsize=(22,12))
sns.countplot(data=df,x='State',order=df['State'].value_counts()[:20].index)
```


The seaborn plot above shows how many orders are ordered by the customer from different states.

It shows the top 20 in that list.

```python

df4 = df.groupby('State')['Profit'].sum()
ax  = df4.plot(kind='bar',figsize=(19,12))
ax.set_ylabel('Profit')
```


As of the states, the most profited state for the spuperstore is California.

This intution can be obtained from the above State vs Profit plot.

Below plot shows the scattering points of the sub-category feature against profit made for each shippment.

```python
fig , ax = plt.subplots(figsize=(22,12))
sns.scatterplot(data=df,x='Sub-Category',y='Profit')
```

Now we are coming to the shippment mode and see how that affects the profit of the superstore on a large scale.

```python

df5 = df.groupby('Ship Mode')['Profit'].sum()
ax = df5.plot(kind='bar',alpha=0.75, rot=0,color=['C4', 'C5', 'C6','C0'],figsize=(12,10))
ax.set_ylabel('Profit')
```

We can clearly see that the Standard Class has obtained the maximum profit of all and it seems obvious due to its quality of shippment.

But there is an insight we need to take care of, the Same Day shippment has produced a very low profit which should actually result in a better profit due to the quick delivery.

Visualizing different regions of the US against Profit.

```python
df6 = df.groupby('Region')['Profit'].sum()
ax = df6.plot(kind='bar',alpha=0.75, rot=0,color=['C2', 'C1', 'C8','C5'],figsize=(12,10))
ax.set_ylabel('Profit')
```

West part of the US has provided highest profit amount to the superstore in the barplot above.

```python
fig , ax = plt.subplots(figsize=(15,10))
sns.heatmap(data=df.corr())
```

Visualizing how the Segment (type of customers) affect the superstore.

```python
fig , ax = plt.subplots(figsize=(10,7))
sns.countplot(data=df,x='Segment',palette='Greens_d')
```

The countplot above shows that the Consumer segment has the highest number of shippment than the Corporate or Home office segments.

```python
df7 = df.groupby('Segment')['Profit'].sum()
ax = df7.plot(kind='bar',alpha=0.75, rot=0,figsize=(12,10))
ax.set_ylabel('Profit')
```

The profit plot also reflects the same as the countplot, that the Consumer segment resulted in more profit than the other two.
